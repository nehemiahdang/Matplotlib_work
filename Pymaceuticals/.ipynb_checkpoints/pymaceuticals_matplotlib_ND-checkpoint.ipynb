## Dependencies and starter code

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress

# Study data files
mouse_metadata = "data/Mouse_metadata.csv"
study_results = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata)
study_results = pd.read_csv(study_results)

# Combine the data into a single dataset
merge_data = pd.merge(mouse_metadata, study_results, how='outer', on='Mouse ID')
merge_data
```

```python
# Pull up the second set of data to understand what's in it
study_results
```

```python
# Stored Variance of each Drug Regimen
Capomulin = merge_data.loc[merge_data["Drug Regimen"] == "Capomulin"]
var1 = np.var(Capomulin["Tumor Volume (mm3)"])

Ceftamin = merge_data.loc[merge_data["Drug Regimen"] == "Ceftamin"]
var2 = np.var(Ceftamin["Tumor Volume (mm3)"])

Infubinol = merge_data.loc[merge_data["Drug Regimen"] == "Infubinol"]
var3 = np.var(Infubinol["Tumor Volume (mm3)"])

Ketapril = merge_data.loc[merge_data["Drug Regimen"] == "Ketapril"]
var4 = np.var(Ketapril["Tumor Volume (mm3)"])

Naftisol = merge_data.loc[merge_data["Drug Regimen"] == "Naftisol"]
var5 = np.var(Naftisol["Tumor Volume (mm3)"])

Placebo = merge_data.loc[merge_data["Drug Regimen"] == "Placebo"]
var6 = np.var(Placebo["Tumor Volume (mm3)"])

Propriva = merge_data.loc[merge_data["Drug Regimen"] == "Propriva"]
var7 = np.var(Propriva["Tumor Volume (mm3)"])

Ramicane = merge_data.loc[merge_data["Drug Regimen"] == "Ramicane"]
var8 = np.var(Ramicane["Tumor Volume (mm3)"])

Stelasyn = merge_data.loc[merge_data["Drug Regimen"] == "Stelasyn"]
var9 = np.var(Stelasyn["Tumor Volume (mm3)"])

Zoniferol = merge_data.loc[merge_data["Drug Regimen"] == "Zoniferol"]
var10 = np.var(Zoniferol["Tumor Volume (mm3)"])

drug_var = [var1,var2,var3,var4,var5,var6,var7,var8,var9,var10]
drug_var
```

```python
# Stored Standard Deviation of each Drug Regimen
std1 = np.std(Capomulin["Tumor Volume (mm3)"])
std2 = np.std(Ceftamin["Tumor Volume (mm3)"])
std3 = np.std(Infubinol["Tumor Volume (mm3)"])
std4 = np.std(Ketapril["Tumor Volume (mm3)"])
std5 = np.std(Naftisol["Tumor Volume (mm3)"])
std6 = np.std(Placebo["Tumor Volume (mm3)"])
std7 = np.std(Propriva["Tumor Volume (mm3)"])
std8 = np.std(Ramicane["Tumor Volume (mm3)"])
std9 = np.std(Stelasyn["Tumor Volume (mm3)"])
std10 = np.std(Zoniferol["Tumor Volume (mm3)"])

drug_std = [std1,std2,std3,std4,std5,std6,std7,std8,std9,std10]
drug_std
```

```python
# Stored SEM of each Drug Regimen
# drug_std[x] / np.sqrt(drug_length[x])
sem1 = drug_std[0] / np.sqrt(drug_length[0])
sem2 = drug_std[1] / np.sqrt(drug_length[1])
sem3 = drug_std[2] / np.sqrt(drug_length[2])
sem4 = drug_std[3] / np.sqrt(drug_length[3])
sem5 = drug_std[4] / np.sqrt(drug_length[4])
sem6 = drug_std[5] / np.sqrt(drug_length[5])
sem7 = drug_std[6] / np.sqrt(drug_length[6])
sem8 = drug_std[7] / np.sqrt(drug_length[7])
sem9 = drug_std[8] / np.sqrt(drug_length[8])
sem10 = drug_std[9] / np.sqrt(drug_length[9])

drug_sem = [sem1,sem2,sem3,sem4,sem5,sem6,sem7,sem8,sem9,sem10]
drug_sem
```

```python
# Merging together the mean and median data
merge_table = pd.merge(drug_mean,drug_median,on="Drug Regimen")
merge_table
```

```python
# Renaming the columns
summary_table = merge_table.rename(columns={
    "Tumor Volume (mm3)_x":"Mean",
    "Tumor Volume (mm3)_y":"Median",
})
summary_table
```

```python
# Displaying all the information into a summary table
summary_table["Variance"] = drug_var
summary_table["Standard Deviation"] = drug_std
summary_table["SEM"] = drug_sem
summary_table
```

```python
# Pulling up the data to see the final timepoint for the treatment
merge_data.head(10)
```

```python
# Extracting only the max values for each mouse id
max_data = merge_data.groupby("Mouse ID").max()
max_data
```

```python
# Resetting the index after getting only the max timepoint treatment for each individual mouse
max_data = max_data.reset_index()
max_data
```

```python
# Calculate the final tumor volume of each mouse across four of the most promising treatment regimens. Calculate the IQR and quantitatively determine if there are any potential outliers. 
# Created separate data sets from each of the 4 most promising treatments
capomulin_final = max_data.loc[max_data["Drug Regimen"] == "Capomulin", ["Mouse ID", "Drug Regimen", "Timepoint","Tumor Volume (mm3)"]]
ramicane_final = max_data.loc[max_data["Drug Regimen"] == "Ramicane", ["Mouse ID", "Drug Regimen", "Timepoint","Tumor Volume (mm3)"]]
infubinol_final = max_data.loc[max_data["Drug Regimen"] == "Infubinol", ["Mouse ID", "Drug Regimen", "Timepoint","Tumor Volume (mm3)"]]
ceftamin_final = max_data.loc[max_data["Drug Regimen"] == "Ceftamin", ["Mouse ID", "Drug Regimen", "Timepoint","Tumor Volume (mm3)"]]
```

```python
# Getting the quartiles of the tumor volumes, lower quartile, upper quartile and interquartile range
# Capomulin 
quartiles = capomulin_final["Tumor Volume (mm3)"].quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq - lowerq

print(f'The Capomulin IQR is {iqr}')

# Getting the upper and lower bounds to get the range to see potential outliers
lower_bound = lowerq - (1.5 * iqr)
upper_bound = upperq + (1.5 * iqr)
print(f'The bounds are between {lower_bound} and {upper_bound}')
print(f'The min is {min(capomulin_final["Tumor Volume (mm3)"])} and max is {max(capomulin_final["Tumor Volume (mm3)"])}')
```

```python
# Ramicane 
quartiles = ramicane_final["Tumor Volume (mm3)"].quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq - lowerq

print(f'The Ramicane IQR is {iqr}')

lower_bound = lowerq - (1.5 * iqr)
upper_bound = upperq + (1.5 * iqr)
print(f'The bounds are between {lower_bound} and {upper_bound}')
print(f'The min is {min(ramicane_final["Tumor Volume (mm3)"])} and max is {max(ramicane_final["Tumor Volume (mm3)"])}')
```

```python
capomulin_data = merge_data.loc[merge_data["Drug Regimen"] == "Capomulin", ["Mouse ID", "Drug Regimen", "Timepoint","Tumor Volume (mm3)"]]
ramicane_data = merge_data.loc[merge_data["Drug Regimen"] == "Ramicane", ["Mouse ID", "Drug Regimen", "Timepoint","Tumor Volume (mm3)"]]
infubinol_data = merge_data.loc[merge_data["Drug Regimen"] == "Infubinol", ["Mouse ID", "Drug Regimen", "Timepoint","Tumor Volume (mm3)"]]
ceftamin_data = merge_data.loc[merge_data["Drug Regimen"] == "Ceftamin", ["Mouse ID", "Drug Regimen", "Timepoint","Tumor Volume (mm3)"]]
```

```python
# Selecting a mouse id from the list and separating it into a new dataset
mouse_id = "s185"
capo_one_mouse = capomulin_data.loc[capomulin_data["Mouse ID"] == mouse_id]
capo_one_mouse
```

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin
capo_time = capo_one_mouse["Timepoint"]
capo_tumor = capo_one_mouse["Tumor Volume (mm3)"]
plt.plot(capo_time, capo_tumor)

plt.title("Timepoint Vs. Tumor Volume for 1 Mouse with Capomulin Treatment")
plt.xlabel("Timepoint")
plt.ylabel("Tumor Volume (mm3)")

plt.grid()
plt.show()
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen
# Created a dataset for this task
capomulin_data2 = merge_data.loc[merge_data["Drug Regimen"] == "Capomulin", ["Mouse ID", "Drug Regimen", "Timepoint", "Tumor Volume (mm3)", "Weight (g)"]]
capomulin_data2
```

```python
# Grouped the data by mouse id to get the data for each mouse
capo_mean = capomulin_data2.groupby('Mouse ID').mean()
capo_mean
```

```python
# Stored the values of weight and average of tumor volume
weight = capo_mean["Weight (g)"]
tumor_avg = capo_mean["Tumor Volume (mm3)"]

plt.scatter(weight, tumor_avg)

plt.title("Mouse Weight Vs. Average Tumor Volume for Capomulin Regimen")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")

plt.show()
```

```python
# Calculate the correlation coefficient and linear regression model for mouse weight and average
# tumor volume for the Capomulin regimen
```

```python
# Correlation coefficient
correlation = st.pearsonr(weight, tumor_avg)
print(f"The correlation is {round(correlation[0],2)}")
print("Therefore, it's a strong correlation.")

# Building line regression
(slope, intercept, rvalue, pvalue, stderr) = linregress(weight, tumor_avg)
regress_values = weight * slope + intercept

# Building the line equation
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

# Plotting the scatter plot
plt.scatter(weight, tumor_avg)

plt.title("Mouse Weight Vs. Average Tumor Volume for Capomulin Regimen")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")

# Plotting the regression line
plt.plot(weight, regress_values,"r-")

# Printing the line equation on the graph
plt.annotate(line_eq, (18,36), fontsize=20, color='green')

plt.show()
```
